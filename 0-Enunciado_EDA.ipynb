Me doy cuenta que las columnas monetarias estan como objetos. Tengo que pasarlas a int para poder modificarlas. Primero extraigo las comas y cualquier signo de dolar (porsiacaso).

```python
# Defino las columnas a limpiar
columns_to_clean = ['FOB U$S', 'FLETE U$S', 'TOTAL KGS NETO']

# CConvierto los valores
for column in columns_to_clean:
    data_total[column] = (data_total[column]
        .str.replace('$', '')    # Remuevo los signos de dolar
        .str.replace(',', '')    # Remuevo las comas
        .str.strip()             # Remmuevo los espacio en blanco
    )
```

Ahora cambio las filas FILETE U$S, FOB U$S Y TOTAL KGS NETO

```python
data_total[['FLETE U$S', 'FOB U$S', 'TOTAL KGS NETO']] = data_total[['FLETE U$S', 'FOB U$S', 'TOTAL KGS NETO']].apply(pd.to_numeric, errors='coerce')
```

```python
data_total.info()
```

```python
# Define columns to process
columns = ['FLETE U$S', 'TOTAL KGS NETO']

for col in columns:
    # Calculate means by RAZON SOCIAL (excluding zeros and NaN)
    means_by_razon = data_total[data_total[col] > 0].groupby('RAZON SOCIAL')[col].mean()
    
    # Calculate overall mean for fallback
    overall_mean = data_total[data_total[col] > 0][col].mean()
    
    # 1. Replace NaN with RAZON SOCIAL means, then overall mean
    data_total[col] = data_total[col].fillna(data_total['RAZON SOCIAL'].map(means_by_razon))
    data_total[col] = data_total[col].fillna(overall_mean)
    
    # 2. Replace zeros with RAZON SOCIAL means, then overall mean
    data_total.loc[data_total[col] == 0, col] = (
        data_total.loc[data_total[col] == 0, 'RAZON SOCIAL']
        .map(means_by_razon)
        .fillna(overall_mean)
    )

# Verify results
for col in columns:
    print(f"\n{col}:")
    print(f"NaN values: {data_total[col].isna().sum()}")
    print(f"Zero values: {(data_total[col] == 0).sum()}")
```

Paso con las variables no numericas. Siendo que es un valor representativo, voy a distribuir los valores del mismo entre 6 paises.

```python
data_total[data_total['PAIS ORIGEN'] == 'NO ESPECIFICADO']
```

```python
# First, let's get all NO ESPECIFICADO records
no_spec_mask = data_total['PAIS ORIGEN'] == 'NO ESPECIFICADO'
total_no_spec = no_spec_mask.sum()

# Create list of target countries
target_countries = ['RUSSIAN FEDERATION', 'CHINA', 'UNITED STATES', 'PERU', 'MALAYSIA', 'INDONESIA']

# Calculate number of records per country (approximately 1/6 each)
records_per_country = total_no_spec // 6

# Create random assignments
random_assignments = np.repeat(target_countries, records_per_country)
# Handle any remaining records by randomly assigning them
remainder = total_no_spec - len(random_assignments)
if remainder > 0:
    random_assignments = np.concatenate([
        random_assignments,
        np.random.choice(target_countries, remainder)
    ])

# Shuffle the assignments
np.random.shuffle(random_assignments)

# Replace NO ESPECIFICADO with random assignments
data_total.loc[no_spec_mask, 'PAIS ORIGEN'] = random_assignments

# Verify the distribution
print("Distribution after changes:")
for country in target_countries:
    count = len(data_total[data_total['PAIS ORIGEN'] == country])
    print(f"{country}: {count} records")

print("\nRemaining 'NO ESPECIFICADO':", 
      len(data_total[data_total['PAIS ORIGEN'] == 'NO ESPECIFICADO']))
```
